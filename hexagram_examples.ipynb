```python
from pydash import flatten
import matplotlib.pyplot as plt

from src.chinese_checkers.geometry.Vector import Vector
from src.chinese_checkers.geometry.Printer import Printer
from src.chinese_checkers.geometry.Hexagram import Hexagram
```

```python
# set up the hexagram
radius = 7
printer = Printer(14, True)
hexagram = Hexagram(radius)
printer.print(hexagram.hexagram_points, [Vector(0,0)])
```

### Sample hexagram

```python
printer.print(hexagram.hexagram_points, [Vector(0,0)])
```

### Sample corner points

```python
printer.print(
    hexagram.hexagram_points,
    flatten(hexagram.hexagram_corner_points.values()),
    hexagram.triangle_grid,
    hexagram.hexagon_points,
)
```

### Sample Passing Custom Axes

```python
fig, axes = plt.subplots(1, figsize=(4, 4))
axes.set_aspect('equal')
printer = Printer()
printer.print(
    hexagram.hexagram_points,
    axes=axes,
    show_plot=False
)
plt.show()
```
